For this project I wanted to see how the survival rate of Glioblastomas differed between four variables. These include sex (male and female), age, site of the tumor, also if radiation and/or surgery was needed. The data is downloaded from the SEER*Stat database.

Glioblastomas are a form of brain cancer that have a very poor prognosis, and typically results in death within six months of not recieving treatment. Knowing this information I wanted to compare the survival rates to all of the factors listed earlier, and the results shocked me. I have showed the results in the form of bar charts because I believe it is easier to read and sort the values for each different catergory.

The first comparisan I looked at was the survival rates between males and females. The results show that there is a relatively equal change of survival and death. Of all recorded cases (both males and females) less than a 20% of them have survived, and over 80% are deceased. 

Next I compared the survival rate of glioblastomas for different age groups. These results showed that children less than one have the greatest change of survival with almost 40% of recorded cases being children who survived in this age group. A close second was the age group 25-29. The group with the least change of survival is people 85+ with over 99% of cases resulting in death.

Looking at the primary site of the tumor, locations including cerebrum, frontal lobe, temporal lobe, parietal lobe, occipital lobe, ventricle, cerebellum, brain stem, overlapping lesion of brain, and brain. These results showed that individuals with a glioblastoma located on the cerebullum have a greater change of survival than other locations. The changes of survival are still slim, less than 25%, but a greater chance than those with the brain being the primary location. 

Lastly I wanted to see if there is a greater survival chance for individuals who recieve radiation, surgery, both, or neither. These results show that people who recieve intraoperative radiation with other radiation before/after surgery and people who recieve intraoperative radiation have a 0% survival rate with 100% of individuals passing away. The method with about 20% of individuals surviving, and also the highest positive outcome, is having surgery both before an after radiation. The next form of treatment with the highest rate of surviving, about 10%, if recieving radiation after surgery. 

To sum up what I learned from this project glioblastomas are a very unfortunate tumor to have with a very small chance for survival. Gender plays a very small to no role, and having this type of tumor being less than one or between 25-29 was showed to have the best chance for curing. The occipital lobe being the primary location showed the best results for surviving as well as recieving surgery both before and after radiation. 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "C:/Users/mosko/OneDrive/Desktop/CaseListing.Glioblastoma.csv"
Glioblastoma = pd.read_csv(file_path)
```

```python
# Survival for males and females

# Select relevant columns and remove rows with missing values
Glioblastoma_survival = Glioblastoma[['Sex', 'Vital status recode (study cutoff used)']].dropna()

# Count survival outcomes grouped by sex
survival_counts = Glioblastoma_survival.groupby(['Sex', 'Vital status recode (study cutoff used)']).size().unstack()

# Calculate the proportion of each outcome within each sex
survival_rate = survival_counts.div(survival_counts.sum(axis=1), axis=0)

# Plot
plt.figure(figsize=(8, 6))
survival_rate.plot(kind='bar', stacked=True, color=['green', 'red'])
plt.title('Survival Rate by Sex')
plt.ylabel('Proportion')
plt.xlabel('Sex')
plt.legend(title='Vital Status', labels=['Alive', 'Dead'])
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Plot for survival by age group

# Select relevant columns and remove rows with missing values
Glioblastoma_survival = Glioblastoma[['Age recode with <1 year olds', 'Vital status recode (study cutoff used)']].dropna()

# Count survival outcomes grouped by sex
survival_counts = Glioblastoma_survival.groupby(['Age recode with <1 year olds', 'Vital status recode (study cutoff used)']).size().unstack()

# Calculate the proportion of each outcome within each sex
survival_rate = survival_counts.div(survival_counts.sum(axis=1), axis=0)

# Plot the results
plt.figure(figsize=(8, 6))
survival_rate.plot(kind='bar', stacked=True, color=['green', 'red'])
plt.title('Survival Rate by Age')
plt.ylabel('Proportion')
plt.xlabel('Age')
plt.tight_layout()
plt.show()
```

```python
# Plot for survival based on the primary site of tumor

# Select relevant columns and remove rows with missing values
Glioblastoma_survival = Glioblastoma[['Primary Site - labeled', 'Vital status recode (study cutoff used)']].dropna()

# Count survival outcomes grouped by sex
survival_counts = Glioblastoma_survival.groupby(['Primary Site - labeled', 'Vital status recode (study cutoff used)']).size().unstack()

# Calculate the proportion of each outcome within each sex
survival_rate = survival_counts.div(survival_counts.sum(axis=1), axis=0)

# Plot the results
plt.figure(figsize=(8, 6))
survival_rate.plot(kind='bar', stacked=True, color=['green', 'red'])
plt.title('Survival Rate by Site of Tumor')
plt.ylabel('Proportion')
plt.xlabel('Site of Tumor')
plt.tight_layout()
plt.show()
```

```python
# Plot for survival based on Radiation and Surgery

# Select relevant columns and remove rows with missing values
Glioblastoma_survival = Glioblastoma[['RX Summ--Surg/Rad Seq', 'Vital status recode (study cutoff used)']].dropna()

# Count survival outcomes grouped by sex
survival_counts = Glioblastoma_survival.groupby(['RX Summ--Surg/Rad Seq', 'Vital status recode (study cutoff used)']).size().unstack()

# Calculate the proportion of each outcome within each sex
survival_rate = survival_counts.div(survival_counts.sum(axis=1), axis=0)

# Plot the results
plt.figure(figsize=(8, 6))
survival_rate.plot(kind='bar', stacked=True, color=['green', 'red'])
plt.title('Survival Rate by Radiation/Surgery Needs')
plt.ylabel('Proportion')
plt.xlabel('Radiation/Surgery')
plt.tight_layout()
plt.show()
```
